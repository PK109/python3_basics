```python
import random
```

```python
def merge(lst0, lst1):
    ret = []
    if len(lst0) + len(lst1) >2:
        print("Merging below lists:")
        print(*lst0)
        print(*lst1)
    while lst0 and lst1:
        if lst0[0] <= lst1[0]:
            ret.append(lst0.pop(0))
        else:
            ret.append(lst1.pop(0))
    ret.extend(lst0)
    ret.extend(lst1)
    print("Merged list:")
    print(*ret, '\n')
    return ret
```

```python
def mergesort(lst):
    if len(lst) <= 1:
        return lst
    print("Array length: {}".format(len(lst)))
    # random to avoid dead loop for special sequence
    r = lst[round(len(lst)/2)]
    left, mid, right = [], [], []
    for i in lst:
        if i < r:
            left.append(i)
        elif i == r:
            mid.append(i)
        else:
            right.append(i)
    if left:
        print("Left side:")
        if len(left)== 1:
            print("THE ONLY:",*left)
        else:
            print(*left)
    else:
        print("\t##Empty left side")
            
        
    print("Mid:")
    print("THE ONLY:",*mid)
    if right:
        print("Right side:")
        if len(right)== 1:
            print("THE ONLY:",*right)
        else:
            print(*right)
    else:
        print("\t##Empty Right side")
    print("\n")
    left = mergesort(left)
    left.extend(mid)
    right = mergesort(right)
    ret = merge(left, right)
    return ret
```

```python
print ("please input length of integer number array")
lst = []

try:
    n = int(input(">"))
except:
    print("Error")

for i in range(1,n+1):
    lst.append(random.randint(1, n**2))
#lst.extend([int(i) for i in n.split()])
print("Origin: ")
print(*lst,sep= ", ")
print ("\n\n")
sorted_list = mergesort(lst)
print ("Sorted: ")
print(*sorted_list, sep= ", ")
```
