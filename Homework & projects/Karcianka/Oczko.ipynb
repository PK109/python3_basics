```python
#Card class
#zawiera kolor, oznaczenie oraz wartość
import random

suits = ('Hearts', 'Diamonds', 'Spades', 'Clubs')
ranks = ('Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Eight', 'Nine', 'Ten', 'Jack', 'Queen', 'King', 'Ace')
values = {'Two':2, 'Three':3, 'Four':4, 'Five':5, 'Six':6, 'Seven':7, 'Eight':8, 
            'Nine':9, 'Ten':10, 'Jack':10, 'Queen':10, 'King':10, 'Ace':11}
```

```python
class Card:
    
    def __init__(self, suit, rank):
        self.suit = suit
        self.rank = rank
        self.value = values[rank]
        
    def __str__(self):
        return self.rank + " of " + self.suit
```

```python
class Deck:
    
    def __init__(self):
        self.all_cards = []
        for suit in suits:
            for rank in ranks:
                self.all_cards.append(Card(suit,rank))
                
    def __str__(self):
        output = ""
        for card in self.all_cards:
            output += str(card) +"\n"
        return output
    
    def shuffle(self):
        random.shuffle(self.all_cards)
    
    def deal_one(self):
        return self.all_cards.pop()
```

```python
class Player:
    
    def __init__(self, name, bot = True, balance = 0):
        self.name = name
        self.balance = balance
        self.all_cards = []
        self.bot = bot
    
    def add_cards(self, new_cards):
        if type(new_cards) == type([]):
            self.all_cards.extend(new_cards)
        else:
            self.all_cards.append(new_cards)
    
    def __str__(self):
        return f'Player {self.name} has {len(self.all_cards)} cards.'
    
    def clear_cards(self):
        self.all_cards = []
        
    def show_cards(self, show_all = False):
        print(str(self))
        if self.bot and show_all == False:
            for card in self.all_cards[:-1]:
                print ("\t" + str(card))
            print("\t** Hidden Card **")
        else:
            for card in self.all_cards:
                print ("\t" + str(card))
        
        if self.bot == False or show_all == True:
            print(f"{self.name} has {self.score()} points.")
        print("\n")
        
    def withdraw(self, ammount):
        if self.bot == True:
            pass
        if self.balance >= ammount and ammount > 0:
            self.balance -= ammount
            print("Current deposit updated to {}\n".format(self.balance))
            return True
        else:
            print("Withdraw unavailable")
            print("Current deposit updated to {}\n".format(self.balance))
            return False
        
    def deposit(self, ammount):
        if self.bot == True:
            pass
        self.balance += ammount
        print("Current deposit updated to {}\n".format(self.balance)) 
        
    def score(self):
        summ = 0
        aces = 0
        for card in self.all_cards:
            summ += card.value
            if card.rank == 'Ace':
                aces += 1
        
        while summ > 21 and aces > 0:
            summ -= 10
            aces -= 1
        
        return summ
    
```
