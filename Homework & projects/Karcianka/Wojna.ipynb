```python
#Card class
#zawiera kolor, oznaczenie oraz wartość
import random

suits = ('Hearts', 'Diamonds', 'Spades', 'Clubs')
ranks = ('Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Eight', 'Nine', 'Ten', 'Jack', 'Queen', 'King', 'Ace')
values = {'Two':2, 'Three':3, 'Four':4, 'Five':5, 'Six':6, 'Seven':7, 'Eight':8, 
            'Nine':9, 'Ten':10, 'Jack':11, 'Queen':12, 'King':13, 'Ace':14}
```

```python
class Card:
    
    def __init__(self, suit, rank):
        self.suit = suit
        self.rank = rank
        self.value = values[rank]
        
    def __str__(self):
        return self.rank + " of " + self.suit
```

```python
class Deck:
    
    def __init__(self):
        self.all_cards = []
        for suit in suits:
            for rank in ranks:
                self.all_cards.append(Card(suit,rank))
                
    def __str__(self):
        output = ""
        for card in self.all_cards:
            output += str(card) +"\n"
        return output
    
    def shuffle(self):
        random.shuffle(self.all_cards)
    
    def deal_one(self):
        return self.all_cards.pop()
```

```python
class Player:
    
    def __init__(self, name):
        self.name = name
        self.all_cards = []
    
    def remove_card(self):
        return self.all_cards.pop()
    
    def add_cards(self, new_cards):
        if type(new_cards) == type([]):
            self.all_cards.extend(new_cards)
        else:
            self.all_cards.append(new_cards)
    
    def __str__(self):
        return f'Player {self.name} has {len(self.all_cards)} cards.'
    
    def shuffle(self):
        random.shuffle(self.all_cards)
    
```

```python
#Game Setup
ania = Player("Ania")
przemek = Player("Przemek")
drawing_ammount = 3
```

```python
#rozdanie kart
deck = Deck()
deck.shuffle()
for x in range(int(len(deck.all_cards)/2)):
    przemek.add_cards(deck.deal_one())
    ania.add_cards(deck.deal_one())
# game:
game_on = True
round_num = 0
cnt_przemek_wins = 0
cnt_ania_wins = 0

while game_on:
    
    round_num += 1
    if round_num % 20 == 0:
        ania.shuffle()
        przemek.shuffle()
        print("Przetasowanie")
    print("Runda " + str(round_num))
    if(len(ania.all_cards) == 0):
        print(ania)
        print(przemek.name + " wins!")
        break
    
    if(len(przemek.all_cards) == 0):
        print(przemek)
        print(ania.name + " wins!")
        break
    
    #start new round
    ania_played_cards = []
    ania_played_cards.append(ania.remove_card())
    
    przemek_played_cards = []
    przemek_played_cards.append(przemek.remove_card())
    
    during_war = True
    while during_war:
        if ania_played_cards[-1].value == przemek_played_cards[-1].value:
            print("War!")
            for _ in range(drawing_ammount):
                if(len(ania.all_cards) > 0):
                    ania_played_cards.append(ania.remove_card())
                else:
                    print(ania.name + " is missing cards.")
                    during_war = False
                if(len(przemek.all_cards) > 0):
                    przemek_played_cards.append(przemek.remove_card())
                else:
                    print(przemek.name + " is missing cards.")
                    during_war = False
        elif ania_played_cards[-1].value > przemek_played_cards[-1].value:
            ania.add_cards(ania_played_cards)
            ania.add_cards(przemek_played_cards)
            during_war = False
            cnt_ania_wins += 1
        else:
            przemek.add_cards(ania_played_cards)
            przemek.add_cards(przemek_played_cards)
            during_war = False
            cnt_przemek_wins += 1
    
print(f"Statistics:\n{ania.name} wins {cnt_ania_wins} rounds.\n{przemek.name} wins {cnt_przemek_wins} rounds")
```
